# Прогнозирование оттока клиентов

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах

Описание услуг:

Оператор предоставляет два основных типа услуг: 
- Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
  
Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).
  
Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек

```python
def camel_to_snake(camel_case):
    """
    Переводит верблюжий регистр в змеиный
    Параметры:
        camel_case: принимает верблюжий текст
    Возвращаемое значение:
        snake_case: змеиный текст
    """
    snake_case = ''
    for i in range(len(camel_case)):
        if camel_case[i].isupper():
            # Если текущая буква заглавная и это не первый символ, и предыдущая буква не заглавная, добавляем подчёркивание
            if i > 0 and not camel_case[i-1].isupper():
                snake_case += '_'
            snake_case += camel_case[i].lower()
        else:
            snake_case += camel_case[i]
    return snake_case
```

```python
def target (end_date):
    """
    Возвращает бинарное значение таргета
    came out - контракт расторгнут - ушел
    stayed - клиент продолжает пользоваться услугами - остался
    """
    if end_date != 'No':
        return 1
    return 0
```

```python
def hist_box_status(df, status_col, y_text):
    '''
    Возвращает гистограмму и boxplot количественного признака в разрезе "ушёл - остался".
    Параметры:
        df: входящий датасет
        status_col: наименование столбца со статусом
        y_text: наименование значения по y
    Возвращаемое значение:
        plt.show(): вывод гистограммы и ящика с усами
    '''
    for feature in con:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))

        # Разбиваем данные по статусу и строим гистограммы
        for status, group in df.groupby(status_col):
            group[feature].plot(kind='hist', bins=10, ax=axes[0], alpha=0.6, label=status)
        axes[0].set_title(f'Признак: {feature} \n Гистограмма распределения')
        axes[0].set_ylabel(f'Количество {y_text}')
        axes[0].set_xlabel(f'{feature}')
        axes[0].legend(title='Статус: 1-ушел, 0-остался')

        # Строим boxplot для каждой группы
        sns.boxplot(x=status_col, y=feature, data=df, ax=axes[1])
        axes[1].set_title(f'Признак: {feature} \n Разброс значений')
        axes[1].set_xlabel('Статус: 1-ушел, 0-остался')
        axes[1].set_ylabel('Значение признака')

        plt.grid(True)
        plt.show()
```

```python
def barh_status(df, feature, status_col, х_text):
    '''
    Возвращает столбчатую диаграмму дискретного или категориального признака в разрезе "ушёл - остался".
    Параметры:
        df: входящий датасет
        feature: признак
        status_col: наименование статусного столбца
        y_text: наименование значения по y
    Возвращаемое значение:
        plt.show(): вывод столбчатой диаграммы
    '''
    # Группируем данные по статусу и признаку
    grouped = df.groupby([status_col, feature]).size().unstack(fill_value=0)

    # Создаём столбчатую диаграмму
    ax = grouped.plot(kind='barh', stacked=False, figsize=(8, 3)) #stacked=True тогда 
                                    # все цвета продолжают друг друга в один столбик
    plt.xlabel(f'Количество {х_text}')
    plt.ylabel('0-остался, 1-ушел')  
    plt.grid(True)
    plt.title(f'Распределение признака: {feature}')
    plt.legend()
    return plt.show()
```

## Подготовка данных

*Для сравнения планирую обучить 4 модели: линейную, дерево и бустинги. У этих моделей разное отношение к подготовке данных (кодированию и масштабированию). Например, для дерева можно использовать OrdinalEncoder даже для неранговых признаков, если не хотим раздувать количество признаков, а вот для линейной модели такой фокус не пройдет. Ей если ранги есть, то OrdinalEncoder, а если нет, то только OneHotEncoder. Бустинг бустингу тоже рознь. Если, например LightGBM, то ей нужен кодировщик как 1му дереву, а если CatBoost, то у него вообще свой встроенный кодировщик. Поэтому готовить данные будем по разному*

### Сформируем наши выборки

*Не забываем про дисбаланс, используем стратификацию*

```python
X = df.drop(['target'], axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)
```

### Подготовим списки признаков

*Категориальные: Вообще-то у нас нет супермультиклассовых признаков (у нас максимум 3 класса), нет ранговых признаков. Поэтому OneHotEncoder нам подходит лучше всего*

```python
ohe_columns = X.select_dtypes(include=['object']).columns.tolist()
```

*Количественные отмасштабируем скейлером. Но для линейной модели будем его использовать, а для деревянной нет. Дерево по другому принимает решения, ему, по большому счету, все равно отмасштабированы данные или нет*

```python
num_columns = X.select_dtypes(include=['float64']).columns.tolist()
```

*Таргет ни кодировать, ни масштабировать не нужно. Он у нас итак бинарный числовой*

### OHE

*Мы знаем, что в наших данных нет пропусков, однако нет гарантий, что их не будет в данных на которых модель будет делать реальные предсказания. Поэтому предусмотрим такой вариант с помощью `SimpleImputer`. Если в данных будут пропуски, он заполнит их константой 'other'*

```python
# SimpleImputer + OHE

ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, 
                                            strategy='constant', 
                                            fill_value='other')),
        ('ohe', OneHotEncoder(drop='first', 
                              handle_unknown='ignore', 
                              sparse=False))
    ]
)
```

### Скейлер для линейной модели

*Возьмем MinMaxScaler. У нас все признаки после кодирования получатся бинарными, поэтому кажется логичным взять MinMaxScaler*

### ColumnTransformer для линейной модели

```python
'''
Аргумент remainder='passthrough' позволит не применять 
шаги пайплайна к данным, которые должны остаться без изменений
'''
preprocessor_log_reg = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

### ColumnTransformer для дерева и LGBMClassifier

```python
preprocessor_tree = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns)     
    ], 
    remainder='passthrough'
)
```

## Обучение моделей

*Искать лучшую модель решила отдельными пайплайнами. Хотя больше люблю все закинуть в один, чтобы меньше кода и локаничнее все выглядело. НО плюс отдельных пайплайнов - можно расширить диапазон подбора гиперпараметров и быстренько потестить, а не ждать пока просчитается ячейка с 3-4 моделями*

*Критерием модели задана ROC-AUC, естественно по ней и ведем подбор*

*У нас нет данных, по которым можно было бы построить бизнес логику для определения порога. Поэтому важность ошибок считаем равнозначной и порог оставляем по умолчанию*

### LogisticRegression

*Учтем веса классов class_weight={1: 0, 1: 6}*

```python
start_time = time.time()
pipe_log_reg = Pipeline(
    [
        ('preprocessor', preprocessor_log_reg),
        ('models', LogisticRegression())
    ]
)
param_grid_log_reg = [
    # словарь для модели LogisticRegression()
    {
        'preprocessor': [preprocessor_log_reg],
        'models': [LogisticRegression(class_weight={1: 0, 1: 6})],
        'models__random_state' : [RANDOM_STATE],
        'models__C': range(1, 500),
        'models__solver': ['liblinear'],
        'models__penalty': ['l1', 'l2']
    }
]
# Не забываем про стратификацию
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rs_log_reg = RandomizedSearchCV(
    pipe_log_reg, 
    param_grid_log_reg, 
    cv=cv,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter = 20
)
rs_log_reg.fit(X_train, y_train)
end_time = time.time()
print(f"Время выполнения: {end_time-start_time} секунд")
print ('Метрика лучшей модели на кроссвалидации:', rs_log_reg.best_score_)
```

```python
result_log_reg = pd.DataFrame(rs_log_reg.cv_results_)
result_log_reg = result_log_reg[['param_models__penalty', 'param_models__C',
                                  'mean_test_score', 'rank_test_score']]
result_log_reg = rename_columns(result_log_reg)
result_log_reg.sort_values('rank_test_score').head(10)
```

*ВЫВОД: Ну перебор особых сдвигов метрики не дает. Требуемое качество не достингуто. И так как мы не ограничены именно этой моделью, попробуем другую*

### DecisionTreeClassifier

*Сразу скажу: потестила и с указанием весов и без, вообще без разницы. class_weight не буду использовать*

```python
start_time = time.time()
pipe_tree = Pipeline(
    [
        ('preprocessor', preprocessor_tree),
        ('models', DecisionTreeClassifier())
    ]
)
param_grid_tree = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier()],
        'preprocessor': [preprocessor_tree],
        'models__random_state' : [RANDOM_STATE],
        'models__max_depth': range(4, 10),
        'models__max_features': range(2, 17),
        'models__min_samples_split': range(2, 6),
        'models__min_samples_leaf': range(2, 6)
    }
]
# Не забываем про стратификацию
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rs_tree = RandomizedSearchCV(
    pipe_tree, 
    param_grid_tree, 
    cv=cv,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter = 50
)
rs_tree.fit(X_train, y_train)
end_time = time.time()
print(f"Время выполнения: {end_time-start_time} секунд")
print ('Метрика лучшей модели на кроссвалидации:', rs_tree.best_score_)
```

```python
result_tree = pd.DataFrame(rs_tree.cv_results_)
result_tree = result_tree[['param_models__min_samples_split', 'param_models__min_samples_leaf',
                           'param_models__max_features', 'param_models__max_depth',
                                  'mean_test_score', 'rank_test_score']]
result_tree = rename_columns(result_tree)
result_tree.sort_values('rank_test_score').head(10)
```

*ВЫВОД: Ну метрика, конечно, получше, но не достаточно. Посмотрим, что нам даст бустинг*

### LGBMClassifier

```python
start_time = time.time()
pipe_lgbm = Pipeline(
    [
        ('preprocessor', preprocessor_tree),
        ('models', LGBMClassifier())
    ]
)
param_grid_lgbm = [
     # словарь для модели LGBMClassifier()
    {
        'models': [LGBMClassifier(objective='binary', verbose=-1)],
        'preprocessor': [preprocessor_tree],
        'models__random_state' : [RANDOM_STATE],
        'models__learning_rate': [0.05, 0.1, 0.3, 0.5, 1],
        'models__n_estimators': range(50, 200),
        'models__max_depth': range(4, 10),
        'models__num_leaves': range(10, 40),  # Максимальное количество листьев в дереве
        'models__min_child_samples': range(18, 22) # Минимальное количество объектов в листьях
    }
]
# Не забываем про стратификацию
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rs_lgbm = RandomizedSearchCV(
    pipe_lgbm, 
    param_grid_lgbm, 
    cv=cv,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter = 50
)
rs_lgbm.fit(X_train, y_train)
end_time = time.time()
print(f"Время выполнения: {end_time-start_time} секунд")
print ('Метрика лучшей модели на кроссвалидации:', rs_lgbm.best_score_)
```

```python
result_lgbm = pd.DataFrame(rs_lgbm.cv_results_)
result_lgbm = result_lgbm[['param_models__num_leaves', 'param_models__n_estimators',
                           'param_models__min_child_samples', 'param_models__max_depth',
                           'param_models__learning_rate', 'mean_test_score', 'rank_test_score']]
result_lgbm = rename_columns(result_lgbm)
result_lgbm.sort_values('rank_test_score').head(10)
```

*ВЫВОД: Ну вот, вся первая десятка отвечает требованиям качества. Бустинг победил. Но посмотрим, что нам даст другой бустинг*

### CatBoostClassifier

```python
start_time = time.time()
pipe_catb = Pipeline(
    [
        ('models', CatBoostClassifier())
    ]
)
param_grid_catb = [
     # словарь для модели CatBoostClassifier()
    {
        'models': [CatBoostClassifier(
            cat_features=ohe_columns,
            logging_level='Silent', # Не выводим НИКАКОЙ промежуточной инфы
            random_seed=RANDOM_STATE,
            loss_function='Logloss',  # Функция потерь, норм для бинарной классификации
            eval_metric='AUC'
            )],
        'models__iterations': [100],
        'models__depth': range(4, 10),
        'models__learning_rate': [0.5, 1],
        'models__l2_leaf_reg': [2]
    }
]
# Не забываем про стратификацию
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rs_catb = RandomizedSearchCV(
    pipe_catb, 
    param_grid_catb, 
    cv=cv,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    n_iter = 10
)
rs_catb.fit(X_train, y_train)
end_time = time.time()
print(f"Время выполнения: {end_time-start_time} секунд")
print ('Метрика лучшей модели на кроссвалидации:', rs_catb.best_score_)
```

Таким образом, если бизнесу важнее охватить всех "уходящих", пусть даже в охват попадут те, кто никуда не собирается, то необходимо повышать Recall. Если мы оставляем данную модель для работы, то это можно сделать за счет понижения порога классификации. Но нужно понимать, что изменение порога улучшит одни показатели, но неизбежно ухудшит другие

Для того, чтобы эффективно определить порог классификации необходимо соотнести планируемые затраты на удержание клиента с предполагаемыми потерями, связанными с его уходом
