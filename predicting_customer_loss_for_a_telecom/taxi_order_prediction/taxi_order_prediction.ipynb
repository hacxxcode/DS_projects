```python
def time_series_plot (df, title):
    '''
    Возвращает график временного ряда
            Параметры:
                    df: входящий датасет 
                    title:'название графика'
            Возвращаемое значение:
                    plt.show(): вывод графика
    '''
    plt.figure(figsize=(15, 5))
    plt.plot(df)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
def roll_mean_ctd_plot (df, window):
    '''
    Возвращает график скользящего срднего и std
            Параметры:
                    df: входящий датасет 
                    window: размер окна
            Возвращаемое значение:
                    plt.show(): вывод графика
    '''
    plt.figure(figsize=(15, 5))
    plt.plot(df.rolling(window).mean(), label='Скользящее среднее')
    plt.plot(df.rolling(window).std(), label='Стандартное отклонение')
    plt.title(f'Скользящие среднее и стандартное отклонение (окно {window})')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
def decomposed_func (df, start, end):
    '''
    Возвращает графики составляющих временного ряда
            Параметры:
                    df: входящий датасет 
                    start: начало интервала 
                    end: конец интервала
                    (интервал в формате ['2018-01-01':'2018-01-15'])
            Возвращаемое значение:
                    plt.tight_layout(): вывод графиков
            
            Чтобы график корректно отобразился, указываем его
            оси ax, равными plt.gca() (англ. get current axis,
            получить текущие оси)
    '''
    decomposed = seasonal_decompose(df)
    
    plt.figure(figsize=(15, 15))
    plt.subplot(411)
    df[start:end].plot(ax=plt.gca())
    plt.title('Time series')
    plt.grid(True)
    plt.subplot(412)
    decomposed.trend[start:end].plot(ax=plt.gca())
    plt.title('Trend')
    plt.grid(True)
    plt.subplot(413)
    decomposed.seasonal[start:end].plot(ax=plt.gca())
    plt.title('Seasonality')
    plt.grid(True)
    plt.subplot(414)
    decomposed.resid[start:end].plot(ax=plt.gca())
    plt.title('Residuals')
    plt.grid(True)
    plt.tight_layout()
```

```python
%%time
# Настройка кросс-валидации для временного ряда
tscv = TimeSeriesSplit(n_splits=4)
rmse_values = []

# Проход по всем складкам кросс-валидации и обучение модели
for train_index, test_index in tscv.split(X_train):
    
    train_data = X_train.iloc[train_index]
    test_data = X_train.iloc[test_index]

    # Инициализация и обучение модели SARIMAX на обучающей выборке
    model = sm.tsa.statespace.SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 24))
    fitted_model = model.fit()

    # Оценка модели на валидационной выборке
    predictions = fitted_model.predict(start=test_index[0], end=test_index[-1])

    # Расчет RMSE
    mse = mean_squared_error(test_data, predictions)
    rmse = np.sqrt(mse)
    rmse_values.append(rmse)

rmse_mean_s = np.mean(rmse_values) 
print(f' RMSE на кросс-валидации: {rmse_mean_s}')
```

В общем и целом, ты права в том, что если на последних отрезках ряда тренд усиливается и появляется больший разброс, то ухудшение метрики на тесте может быть связано именно с этим, а не с переобучением модели. Существует несколько этапов, как это обычно можно проверять:
        
* **1) Визуальный анализ предсказаний на тесте** (строим график с фактическими значениями на всем ряду, предсказаниями модели на тренировочной части, предсказаниями модели на тестовой части). Если модель переобучена, то на тренировочной части ее предсказания будут почти идеально повторять факт, а на тестовой части предсказания будут качественно плохими с самого начала — пойдут не туда, будут иметь другую форму или не уловят явный, но новый тренд. Если данные изменились, то модель на тесте может качественно правильно повторять форму тренда или сезонности, но количественно ошибаться из-за возросшей волатильности или из-за того, что новый тренд просто сильнее/слабее, чем был в истории. Предсказания будут "плыть" вместе с данными, но с постоянным отставанием или ошибкой.
       
* **2) Анализ ошибок на тренировочной и тестовой части** (смотрим не только на итоговое значение метрики, но и на ее динамику). При переобучении ошибка на трейне будет систематически и значительно ниже, чем на тесте. При изменении данных ошибка на трейне и на тесте может быть сопоставима по величине в начале тестового периода. Но она будет возрастать со временем внутри тестового отрезка, так как модель, обученная на старых данных, все хуже и хуже описывает новую реальность.

* **3) Использование скользящего окна/расширяющегося окна.** Если на всех фолдах кросс-валидации метрика стабильно ухудшается на последних отрезках — виноваты данные. Если метрика резко и скачкообразно ухудшается на конкретном фолде — возможно, модель переобучилась на каком-то специфическом паттерне.
        
Анализ же остатков необходим, но не достаточен для диагностики переобучения. Он показывает, нет ли в модели недообучения, но не может уверенно показать переобучение.
